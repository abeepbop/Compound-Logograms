```python
# how to write compound characters?
    # the bounds of a component must be defined.
# or, find a new way to define it recursively...
# writing it (the instructions) is different than /storing/ it. 
import pygame
import numpy as np
from logogram import *

bg_color = (249,229,200)
t = 0

pygame.init()
clock = pygame.time.Clock()
size = width, height = 700, 700

screen = pygame.display.set_mode(size)

logograms = []
num_logs = 0

num_strokes = 15 #zero indexed, including nothing
increment = width / num_strokes

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT: sys.exit()
            
    keys = pygame.key.get_pressed()
    if keys[pygame.K_q]:
        pygame.display.quit()
        pygame.quit()
        #sys.exit()
        exit()  
    
    offset = 70
    
    pos1 = [width//2-offset,height//2]
    radius = 100
    
    top = Logogram([2,[3,4,5]], pos1)
    top.radius = radius
    top.center[1] = top.center[1] - radius
    logograms.append(top)
    mid = Logogram([0], pos1)
    mid.radius = radius
    logograms.append(mid)
    cross = Logogram([2,[3]], pos1)
    cross.radius = radius
    cross.center[1] = cross.center[1]+radius
    logograms.append(cross)
    
    #it's important for the three components to be drawn 
    #s.t. the whole symbol is 2 radius high max and 2 widths high max.
    #in this case, 3 components * 2 radiuses = 6 radiuses
    
    pos2 = [width//2+offset,height//2]
    radius = 100 #start with the width of a regular simple character (like just a circle)
    
    small = []
    top = Logogram([2,[3,4,5]], pos2)
    small.append(top)
    mid = Logogram([0], pos2)
    small.append(mid)
    cross = Logogram([2,[3]], pos2)
    small.append(cross)
    
    radius = radius//len(small)
    for i in range(len(small)):
        component = small[i]
        component.radius = radius
        component.position = [width//2+offset,(height//2)-radius*2]

    #RENDERING STAGE
    screen.fill(bg_color)
    #blit every part of the sunflower.
    for logogram in logograms:
        logogram.draw(screen,bg_color)
    for component in small:
        component.draw(screen,bg_color)
    pygame.display.flip()
    
    dt = clock.tick(60) / 1000
    t += dt
```
